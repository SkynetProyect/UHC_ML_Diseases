<h1> DESCRIPTION </h1>


This database contains 76 attributes, but all published experiments refer to using a subset of 14 of them.  In particular, the Cleveland database is the only one<br> that has been used by ML researchers to date.  The "goal" field refers to the presence of heart disease in the patient.<br>  It is integer valued from 0 (no presence) to 4. Experiments with the Cleveland database have concentrated on simply attempting to distinguish presence (values 1,2,3,4) from absence (value 0).  
<br>
The names and social security numbers of the patients were recently removed from the database, replaced with dummy values.
<br>
One file has been "processed", that one containing the Cleveland database.  All four unprocessed files also exist in this directory.
